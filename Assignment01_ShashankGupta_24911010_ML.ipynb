```python
df1 = df[['Physical Activity', 'Cholesterol Levels']]
mean = df1.groupby('Physical Activity').mean()
print(mean)
plt.bar(mean.index, mean['Cholesterol Levels'])
plt.xlabel('Physical Activity')
plt.ylabel('Average Cholesterol Level')
plt.title('Physical Activity Distribution')
plt.show()
data1 = df1['Cholesterol Levels'][df1['Physical Activity'] == 'High']
data2 = df1['Cholesterol Levels'][df1['Physical Activity'] == 'Moderate']
data3 = df1['Cholesterol Levels'][df1['Physical Activity'] == 'Low']
data = [data1, data2, data3]
labels = ['High', 'Moderate', 'Low']
plt.boxplot(data, labels=labels)
plt.xlabel('Physical Activity')
plt.ylabel('Cholesterol Levels')
plt.title('Cholesterol Levels Distribution by Physical Activity')
plt.show()
```

```python
bad_habbit = df.groupby(['Smoking Status', 'Alcohol Consumption'])['Smoking Status'].count().unstack()
bad_habbit.plot(kind = 'bar', stacked = True)
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.title('Smoking Status and Alcohol Consumption')
plt.show()
```
